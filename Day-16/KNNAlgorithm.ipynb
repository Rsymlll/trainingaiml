## KNN algorithm in supervised machine learning
## KNeighboursClassifier of sklearn library for this one

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = np.array([
    [1,3500], #unfit
    [1,3200], #unfit
    [2,3500], #unfit
    [3,3000], #unfit
    [4,2500], #fit
    [5,2300], #fit
    [6,2000], #fit
    [7,2200], #fit
    [2,3700], #unfit
])
y=['Unfit','Unfit','Unfit','Unfit','Fit','Fit','Fit','Fit','Unfit',]
```

```python
#create the model
model=KNeighborsClassifier(n_neighbors=3)
```

```python
#train the model
model.fit(X,y)
```

```python
#predict
new_person=[[4.5,2600]] #4.5 hrs exercise, 2600 calories
pred=model.predict(new_person)
print('Predicted Result:',pred[0])
```

```python
#prediction from taking the values from the users
exercise_hours=float(input('Enter number of exercise hours'))
cal_intake=float(input('Enter number of calories consumed'))
pred_fu=model.predict([[exercise_hours,cal_intake]])
print(f'Exercise Hours {exercise_hours}, Calories Intake {cal_intake} Your are likely {pred_fu[0]}')
```

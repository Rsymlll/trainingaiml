#Supervised machine learning
## linear regression

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
data = pd.read_csv("sales_data.csv")
data.head()
```

```python
x=data[['Advertising_Spend']]
y=data['Sales_Revenue']
```

```python
model=LinearRegression()
model.fit(x,y)
print('Traing Done')
```

```python
new_spend=float(input('Enter your spend to predict sales revenue'))
predicted_revenue=model.predict([[new_spend]])
print(f'New spend {new_spend} Predicted revenue{predicted_revenue}')
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
data = pd.read_csv("sales_data.csv") ##reading the data
data.head(2)
```

```python
##splitting the data
advertisingSpend=data[['Advertising_Spend']]
salesRevenue=data['Revenue_Spend']
```

```python
model=LinearRegression() ##model creating
model.fit(advertisingSpend,salesRevenue) ## train the model
```

```python
prValues=model.predict([[4000]]) ##predict the model
print(prValues)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
data = pd.read_csv("house_price.csv") ##reading the data
data
```

```python
model=LinearRegression() ##model creating
model.fit(x,y)
```

## 2.logistic regression

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np
## collect the data
df = pd.read_csv("pass_fail.csv") ##reading the data
df.head()
```

```python
## split the data
x= df[['Hours']]
y= df[['Result']]
```

```python
## create the data
model=LogisticRegression()
```

```python
model.fit(x,y)
```

```python
hours_study=int(input('Number of studied'))
pred_value=model.predict([[hours_study]])
print(f'After {hours_study} hours study you are likely to {pred_value[0]}')
```

```python
## import all libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder   

## collect the data
df = pd.read_csv("customer_purchase.csv")  # reading the data
df.head()

## convert categorical labels into numbers
le = LabelEncoder()
df['Purchased'] = le.fit_transform(df['Purchased'])

## select input (features) and output (target)
x = df[['Age', 'Salary']]   
y = df['Purchased']         
model=LogisticRegression()
```

```python
model.fit(x,y)
```

```python
new_cust_age=int(input('Enter age'))
new_cust_sal=int(input('Enter salary'))
new_cust=[[new_cust_age,new_cust_sal]]
prediction=model.predict(new_cust)
if (prediction):
    print('Purchased')
else:
    print('Not purchased')
```

```python
## import all libraries
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("spam_dataset.csv")
df
```

```python
x=df['message_content']
y=df['is_spam']
```

```python
#steo 5 : convert text to numbers
cv=CountVectorizer()
x_cv=cv.fit_transform(x)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_cv,y)
```

```python
sample=['Hi Joe,','Hi Joe,']
pred=model.predict(cv.transform(sample))
if(pred.any()):
    print('Not spam')
else:
    print('Spam')
```
